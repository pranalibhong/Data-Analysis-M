```python


def Iqr_capping(df, cols ):
    for col in cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)

        IQR = Q3 - Q1

        # Define outlier thresholds with a higher multiplier
        lower_threshold = Q1- 1.5 * IQR
        upper_threshold = Q3 + 1.5 * IQR
        df[col] = np.where(df[col]>upper_threshold ,upper_threshold, np.where(df[col]  <lower_threshold,lower_threshold,df[col]))




```

```python
# Convert 'Purchase_Date' column to datetime
df_cap['Purchase_Date'] = pd.to_datetime(df_cap['Purchase_Date'])

# Extract year from the 'Purchase_Date' column
df_cap['Year'] = df_cap['Purchase_Date'].dt.year

# Calculate discounted price before and after applying the discount
df_cap['Discounted_Price_Before'] = df_cap['Price_INR']  # Initially assume discounted price is the same as the original price
df_cap['Discounted_Price_After'] = df_cap['Price_INR'] - df_cap['Discount_INR']

# Calculate total sales before and after applying the discount
df_cap['Total_Sales_Before_Discount'] = df_cap['Price_INR'] * df_cap['Quantity_Purchased']
df_cap['Total_Sales_After_Discount'] = df_cap['Discounted_Price_After'] * df_cap['Quantity_Purchased']

# Group by year and calculate total sales before and after
total_sales_by_year_before = df_cap.groupby('Year')['Total_Sales_Before_Discount'].sum()
total_sales_by_year_after = df_cap.groupby('Year')['Total_Sales_After_Discount'].sum()

# Plotting
plt.figure(figsize=(12, 6))

# Plot total sales before discount
plt.plot(total_sales_by_year_before.index, total_sales_by_year_before.values, marker='o', linestyle='-', color='b', label='Sales Before Discount')

# Plot total sales after discount
plt.plot(total_sales_by_year_after.index, total_sales_by_year_after.values, marker='o', linestyle='-', color='r', label='Sales After Discount')

plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.title('Impact of Discounts on Total Sales Over Years')
plt.legend()
plt.grid(True)
plt.show()
```

# Brand Performance

### Q1 : 	Determine the market share of each brand in different product categories.

```python
# Step 1: Group by Product_Brand and Product_Name, and calculate total sales
brand_sales = df_cap.groupby(['Product_Brand', 'Product_Name'])['Total_Sales_Generated'].sum().reset_index()

# Step 2: Calculate market share for each brand in each product category
brand_market_share = brand_sales.groupby('Product_Name')['Total_Sales_Generated'].apply(lambda x: x / x.sum()).reset_index()
brand_market_share.columns = ['Product_Name', 'Product_Brand', 'Market_Share']

# Step 3: Pivot the data to get market share for each brand in each product category
market_share_pivot = brand_market_share.pivot_table(index='Product_Name', columns='Product_Brand', values='Market_Share', aggfunc='sum')
# Validate data
assert (market_share_pivot >= 0).all().all(), "Market share values cannot be negative"
assert market_share_pivot.notnull().all().all(), "Market share values cannot be null"

# Visualize data
import seaborn as sns
sns.set()
sns.heatmap(market_share_pivot, annot=True, cmap='coolwarm')
```
